```python
import numpy as np
from numpy import linalg as LA

f = open("glove.6B.50d.txt", "r")
data = f.readlines()
i = 0
word = []
words = {}
Embeds = np.empty((len(data), len(data[0].split(' ')) - 1), dtype=np.double)
for line in data:
    splitted = line.split(" ")
    words[i] = splitted[0]
    Embeds[i] = (splitted[1:])
    i = i + 1

Embeds = Embeds / (LA.norm(Embeds, axis = 1).reshape(400000,1))
```

```python
word = np.array(word)
i = np.where(word == "cool") [0][0]
scal = np.matmul(Embeds[i], Embeds.T)
J = np.argsort(scal)

s = list(word[J[-26:]])
s.reverse()
maxes = scal[J[-26:]]
maxes = (np.abs(2 - 2*maxes))**(0.5)
maxes = list(maxes)
maxes.reverse()
s = list(map(str,s))
list2Str = ""
for k in range(26):
    list2Str = list2Str + s[k] + "[" + str(round(maxes[k],5)) + "] "
list2Str
```

```python
i = 0
max_vals = []
for i in range(4000):
    scal = np.matmul(Embeds[i*100:(i+1)*100], Embeds.T)
    scal = np.triu(scal, i*100+1)
    scal = np.ravel(scal)
    J = np.argpartition(scal, (-25, -1))[-25:]
    J_local = np.array(list(zip(J//400000 + i * 100, J%400000)))
    current_vals = scal[J]
    J_local = np.flip(J_local, axis=0)
    current_vals = np.flip(current_vals,axis=0)
#     print(J_local)
#     print(current_vals)
    if i==0:
        max_ind = J_local
    else:
        max_ind = np.append(max_ind, J_local, axis=0)
        
    max_vals = np.append(max_vals, current_vals)
    J_new = np.argpartition(max_vals, (-25, -1))[-25:]
    max_vals = max_vals[J_new]
    max_ind = max_ind[J_new]
    
print(max_ind)
J_new = np.argsort(max_vals)
J_new = np.flip(J_new, axis=0)
max_vals = max_vals[J_new]
max_ind = max_ind[J_new]
for i in range(25):
    print('(' + words[max_ind[i,0]] + ', ' + words[max_ind[i,1]] + ' | ' + str(max_vals[i]) + ')')
```
